```python
def winner():

  player1 = input("player1: please enter rock, paper, scissors ").strip().lower()
  player2 = input("player2: please enter rock, paper, scissors ").strip().lower()

  if player1 == player2:
     print("tie")

  elif player1 == "rock" and player2 == "scissors":
     print("player1 won")

  elif player1 == "rock" and player2 == "paper":
    print("player2 won")

  elif player1 == "paper" and player2 == "rock":
    print("player1 won")

  elif player1 == "paper" and player2 == 'scissors':
    print("player2 won")

  elif player1 == "scissors" and player2 == "paper":
    print("player1 won")
  elif player1 == "scissors" and player2 == "rock":
    print("player2 won")
  elif player1 == "scissors" and player2 == "scissors":
    print("tie")
  else:
    print("have a good day")

winner()
```
