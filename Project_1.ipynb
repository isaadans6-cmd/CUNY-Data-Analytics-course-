```python
unique_years = df['Year'].unique()
unique_school = df['School Name'].unique()
school_level = df['School Level'].unique()
subject = df['Regents Exam'].unique()

print(
    f"The raw dataset contains regent scores of {len(unique_school)} schools. "
    f"School level ranges are {school_level}, for the academic years {unique_years}. "
    f"Subject includes {subject}. Example of some schools includes: {unique_school[:3]}"
)

benchmark_school ='Bronx Community High School'
print(f"""For the following analysis, we will compare regents performance of {benchmark_school} to that of the citywide high schools for 2015-2017.""")
print()

high_school_df = df[df['School Level']=='High school'].copy()
bronx = df[df['School DBN'].str.contains('x', case=False, na=False)]
```

```python
# Convert 'Mean Score' to numeric, coercing 's' to NaN
high_school_df['Mean Score'] = pd.to_numeric(high_school_df['Mean Score'], errors='coerce')

# Filter for the benchmark school
benchmark_school_df = high_school_df[high_school_df['School Name'] == benchmark_school]

# Calculate citywide mean scores for high schools
citywide_mean_scores = high_school_df.groupby(['Regents Exam', 'Year'])['Mean Score'].mean().reset_index()
citywide_mean_scores['School Name'] = 'Citywide Average'

# Combine benchmark school data and citywide average data
comparison_df = pd.concat([benchmark_school_df, citywide_mean_scores], ignore_index=True)

# Display the combined data for verification
print("Comparison DataFrame head:")
print(comparison_df.head())
```

```python
import matplotlib.pyplot as plt

# Filter out NaN values from Mean Score for plotting
plot_df = comparison_df.dropna(subset=['Mean Score'])

# Create the bar chart
plt.figure(figsize=(15, 8))
sns.barplot(
    data=plot_df,
    x='Regents Exam',
    y='Mean Score',
    hue='School Name',
    errorbar=None,
    palette={'Bronx Community High School': 'skyblue', 'Citywide Average': 'lightcoral'}
)

plt.title(f'Mean Regents Exam Scores: {benchmark_school} vs. Citywide High School Average (2015-2017)')
plt.xlabel('Regents Exam')
plt.ylabel('Mean Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='School Type')
plt.tight_layout()
plt.show()
```
